## grafikleri görselleştirmesi için sadece jupiter notebookta çalışır

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,90,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"r") # yatay ilki x, ikinci y , 3. renk
plt.xlabel("Yas")  ## x in adı
plt.ylabel("Kilo")  ## y nin adı
plt.title("change of weight as age")  ## başlık

## başka editörde plt.show() yazılmalı
```

## figure oluşturma boş bir kanvas oluşturma 

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2,0.2,0.9,0.9])  ## boyut
figureAxes.plot(numpyDizi1,numpyDizi2,"g")  ## shitf artı tab özellik gösterme
figureAxes.set_ylabel("Y Ekseni Veri Ismi")
figureAxes.set_xlabel("X Ekseni Veri Ismi")
figureAxes.set_title("Grafik Basligi")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])  ## ilk 2 nokta grafiklerin öbür grafiğe göre nerede durduğunu belirtiyor
eksen1.plot(numpyDizi1,numpyDizi2,"r")
eksen1.set_xlabel("x")
eksen1.set_ylabel("y")
eksen1.set_title("Ana Figur")

eksen2.plot(numpyDizi1,numpyDizi2,"b")
eksen2.set_xlabel("x")
eksen2.set_ylabel("y")
eksen2.set_title("Diger Figur")
```
